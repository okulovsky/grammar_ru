```python
import zipfile
from pathlib import Path
from tg.grammar_ru.corpus import CorpusBuilder
from tg.grammar_ru.features import PyMorphyFeaturizer, SlovnetFeaturizer
```

```python
retell_dir_name = ['gp','game_of_t','witcher','Chekhov','all_retell']
title_names = ["Harry Potter","Martin","Witcher","Chekhov","All retell"]
retell_paths = [Path(f"../retell_data/{dir_name}") for dir_name  in retell_dir_name]
```

```python
for title_name,retell_path in zip(title_names,retell_paths):
    base_corpus_path = Path(f"./corpuses/{title_name}.retell.corpus.zip")
    CorpusBuilder.convert_interformat_folder_to_corpus(
        base_corpus_path,
        retell_path,
        ['author','book_name','text_type']
    )
    CorpusBuilder.featurize_corpus(
        base_corpus_path,
        Path(f"./featurized_corpuses/{title_name}.featurized.retell.corpus.zip"),
        [
            PyMorphyFeaturizer(),
            SlovnetFeaturizer()
        ]
    )
```

```python
with zipfile.ZipFile(Path("./corpuses/All retell.retell.corpus.zip"), 'a') as zip_file:
    source_path = Path("./to_find_nedded_books/mapping.parquet")
    destination = Path("mapping.parquet")
    zip_file.write(source_path, destination)
with zipfile.ZipFile(Path("./featurized_corpuses/All retell.featurized.retell.corpus.zip"), 'a') as zip_file:
    source_path = Path("./to_find_nedded_books/mapping.parquet")
    destination = Path("mapping.parquet")
    zip_file.write(source_path, destination)
```
