```python
import zipfile
from pathlib import Path
from tg.grammar_ru.corpus import CorpusBuilder
from tg.grammar_ru.features import SnowballFeaturizer
```

```python
title_dir_name = ['eng_game_of_t']
title_names = ["Martin"]
retell_paths = [Path(f"../retell_data/{dir_name}") for dir_name in title_dir_name]
books_path = [Path(f"../book_data/eng_martin/md")]
```

```python
for title_name,retell_path in zip(title_names,retell_paths):
    base_corpus_path = Path(f"../corpuses/eng/{title_name}.retell.corpus.zip")
    CorpusBuilder.convert_interformat_folder_to_corpus(
        base_corpus_path,
        retell_path,
        ['author','book_name','text_type']
    )
    CorpusBuilder.featurize_corpus(
        base_corpus_path,
        Path(f"../featurized_corpuses/eng/{title_name}.featurized.retell.corpus.zip"),
        [
            SnowballFeaturizer(language='eng')
        ]
    )
```

```python
for title_name,book_path in zip(title_names,books_path):
    base_corpus_path = Path(f"../corpuses/books/eng/{title_name}.books.corpus.zip")
    CorpusBuilder.convert_interformat_folder_to_corpus(
        base_corpus_path,
        book_path,
        ['author','book_name','text_type']
    )
    CorpusBuilder.featurize_corpus(
        base_corpus_path,
        Path(f"../featurized_corpuses/books/eng/{title_name}.featurized.books.corpus.zip"),
        [
            SnowballFeaturizer(language='eng')
        ]
    )
```

```python
mapping_path = Path("../mapping/mapping_eng.parquet")
with zipfile.ZipFile(Path("../corpuses/eng/Martin.retell.corpus.zip"), 'a') as zip_file:
    destination = Path("mapping.parquet")
    zip_file.write(mapping_path, destination)
with zipfile.ZipFile(Path("../corpuses/books/eng/Martin.books.corpus.zip"), 'a') as zip_file:
    destination = Path("mapping.parquet")
    zip_file.write(mapping_path, destination)
```

```python
with zipfile.ZipFile(Path("../featurized_corpuses/eng/Martin.featurized.retell.corpus.zip"), 'a') as zip_file:
    destination = Path("mapping.parquet")
    zip_file.write(mapping_path, destination)
with zipfile.ZipFile(Path("../featurized_corpuses/books/eng/Martin.featurized.books.corpus.zip"), 'a') as zip_file:
    destination = Path("mapping.parquet")
    zip_file.write(mapping_path, destination)
```
