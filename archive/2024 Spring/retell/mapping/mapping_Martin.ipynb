```python
def get_mapping(retell_martin_df_,mapping):
    for book_name in retell_martin_df_['book_name'].unique():
        retell_martin_book = retell_martin_df_[retell_martin_df_['book_name'] == book_name]
        book_name = book_name.strip('(роман)').strip()
        martin_book = martin_df[martin_df['header_0'].str.lower().str.contains(book_name.lower())]

        start_retell_chapter, start_mapping_index = retell_martin_book['headers'].to_list()[0], 0
        book_chapters = [chapter.split('/')[-1].strip() for chapter in martin_book['headers'].to_list()]
        while book_chapters[start_mapping_index].lower() not in start_retell_chapter.lower():
            start_mapping_index += 1
        for retell_chapter_file_id, book_chapter_file_id in zip(retell_martin_book.index,martin_book.index[start_mapping_index:]):
            mapping[retell_chapter_file_id] = book_chapter_file_id

mapping = pd.Series()
get_mapping(retell_martin_df[retell_martin_df['text_type'] == 'retell'],mapping)
get_mapping(retell_martin_df[retell_martin_df['text_type'] == 'summary'],mapping)
```

```python
pd.DataFrame(mapping,columns=['id_to']).to_parquet(Path("./mapping.parquet"),compression=None)
```
