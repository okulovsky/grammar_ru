## Глобальная задача: поиск антецедентов местоимения

Чтобы научиться находить в тексте антецеденты местоимений, нужно иметь тексты с известными антецедентами. Работа в этом направлении и заняла три спринта. 

Что было было известно к началу третьего:
- Использование словаря биграмм на парах (родитель местоимения, кандидат в антецеденты) давало очень большое количество ошибок, сложно придумать способы улучшить его применение
- Global vectors значительно уменьшили количество ошибок, хотя их число ещё было близко к половине на многострадальном примере про Лилию

Этот шаг к успеху и мотивировал попытки улучшить отбор данных, на которых можно учиться искать антецеденты.
____

Был ли примерный план действий? Нет, всё началось с осознания, что простой максимум от произведения векторов пары (родитель местоимения, кандидат) при ближайшем рассмотрении не так уж и хорош: вместо Лилии тревога таится в тревоге, бьётся сердце боли и т.п.

1) Для начала был предпринят оверинжиниринг замены имён собственных на другие часто встречающиеся слова (здесь и позднее выяснилось, что slovnet нередко ошибается и помогает загрязнять выборку)
 

Что было с прошлого спринта: каждый кандидат подставлялся на место местоимения, склонялся и вычислялось его произведение с синтаксическим родителем местоимения.

2) Был подготовлен код для выбора только тех пар, чьё произведение значительно больше произведений с другими кандидатами для данного местоимения (сравнение со вторым максимальным и сравнение со средним)

3) В версии кода с предыдущего спринта в кандидатах могли оказаться другие местоимения. Это запутывает связи. Кроме того прилагательные больше способствуют появлению ошибок чем нахождению красивых правильных пар, в конечном счёте в кандидаты отбираются только слова, опознанные slovnet'ом как NOUN или PROPN.

4) Количество возможных кандидатов сокращено до пяти. Да, были красивые случаи правильного выбора более далёких референсов, но ошибок было больше.

5) Теперь в составлении рейтинга кандидатов участвуют произведения с ближайшими соседями, а не только с родителем из синтаксического дерева (определение родителя так же периодически бывает неправильным)
