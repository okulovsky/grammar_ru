# Metrics demo

```python
import numpy as np
from tg.projects.retell.retell_utils.metrics import get_jaccard_index, get_cosine_sim, plot_confint, plot_mutiple_confints, \
    show_statistics_and_bar
from tg.projects.retell.demos._data._read_write import read_retell
from pathlib import Path
```

## Для начала - текст

```python
paths = [Path("../_data/_human_summary/Martin"),Path("../_data/_machine_summary/Martin")]
true_retell,extract_retell = [read_retell(path)[:-1] for path in paths]    
```

## Вычисление метрик сходства двух отрывков текста:
- Косинусное расстояние
- Индекс Жаккара

Метрики вычисляются для пары заданных строк.
В этом примере это машинный и человеческие пересказы.

```python
jaccard_sim = np.array([get_jaccard_index(extract_retell[i],true_retell[i]) for i in range(len(true_retell))])
cos_sim = np.array([get_cosine_sim(extract_retell[i],true_retell[i]) for i in range(len(true_retell))])
print(jaccard_sim[:3],cos_sim[:3],sep='\n')
```

## Информативный вывод значений метрик
#### Для удобства и информативности сущетсвует метод show_statistics_and_bar.
 Он отрисовывает гистограмму для значений метрик, посчитывает и выводит медиану, максимум, минимум метрики.

```python
show_statistics_and_bar(jaccard_sim,cos_sim)
```

## Доверительные интервалы.
#### plot_confint - это обёртка над grbar, позволяет для значений метрик вывести их доверительные интервалы.

Функция делает бутсрап из 1000 элементов, вычисляя позже доверительный интервал.
Аргументом функции можно передать желаемую ориентацию графика.

```python
plot_confint(jaccard_sim=jaccard_sim,cos_sim=cos_sim,orient='h')
plot_confint(jaccard_sim=jaccard_sim,cos_sim=cos_sim,orient='v')
```

## Несколько доверительных интервалов
#### Если есть необходимость изобразить несколько доверительных интервалов, то можно воспользоваться plot_mutiple_confints.
Аргументы функции практически такие же, но теперь необходимо передавать двумерные List, для каждого можно указать ориентацию, а так же общий размер фигуры.

```python
plot_mutiple_confints(jaccard_sims=[jaccard_sim[:5],jaccard_sim[5:10]],
                      cos_sims=[cos_sim[:5],cos_sim[5:10]],
                      orients=['h' for _ in range(2)],
                      figsize=(10,5))
```
