```python
def construct_sentence(
        sentence_id: int,
        cur_narrative_chunk_len: int,
        frame: pd.DataFrame
) -> Tuple[bool, List[str]]:

    dialog_flag = False
    cur_sentence = []
    for word_id in frame[frame['sentence_id'] == sentence_id]['word_id'].unique().tolist():
        cur_sentence.append(
            frame[frame['word_id'] == word_id]['word'].values[0] +
            ' ' * frame[frame['word_id'] == word_id]['word_tail'].values[0]
        )

        if frame[frame['word_id'] == word_id]['word'].values[0] == '\u201c':
            dialog_flag = True

        if frame[frame['word_id'] == word_id]['word'].values[0] == '\u201d':
            dialog_flag = False
            cur_narrative_chunk_len -= len(cur_sentence)
            cur_sentence = []
            break

    return dialog_flag, cur_sentence
```

```python
def add_sentences(
    frame:pd.DataFrame,
    paragraph_id: int
):
    global cur_narrative_chunk_len
    for sentence_id in frame[frame['paragraph_id'] == paragraph_id]['sentence_id'].unique().tolist():
        (dialog_flag, cur_sentence) = construct_sentence(
            sentence_id, cur_narrative_chunk_len, frame)
        if dialog_flag:
            cur_narrative_chunk_len -= len(cur_sentence)
        else:
            cur_narrative_chunk.append(''.join(cur_sentence))
```

```python
with open(Path('./tasks/task_narrative.json'), 'a', encoding='utf-8') as file_output:
    for frame in martin_reader.get_frames():
        cur_frame_num += 1
        print(
            f"Processing frame with id: {frame['file_id'].values[1]}, {cur_frame_num}/{frames_num}",
            end="\r"
        )
        dialog_flag = False
        prev_narrative = False
        for paragraph_id in frame['paragraph_id'].unique().tolist():
            paragraph_length = frame[frame['paragraph_id'] == paragraph_id].count().values[1]
            cur_paragraph = frame[frame['paragraph_id'] == paragraph_id]
            if cur_narrative_chunk_len + paragraph_length > max_words:
                file_output.write(
                    json.dumps({
                            "id": uuid.uuid1(),
                            "file_id": frame["file_id"],
                            "text_narrative": ''.join(cur_narrative_chunk),
                            "text_dialog": "",
                            "retell": "",
                            "word_start_id": 0,
                            "word_end_id": 0
                    }) + ','
                )
                cur_narrative_chunk_len = 0
                cur_narrative_chunk = []

            add_sentences(frame, paragraph_id)

            cur_narrative_chunk.append('\n')
            cur_narrative_chunk_len += paragraph_length
```
