```python
from grammar_ru.corpus.corpus_reader import CorpusReader
from grammar_ru.corpus.corpus_writer import CorpusWriter
from pathlib import Path
```

```python
ru_books_reader = CorpusReader(Path('./files/books.base.zip'))

books_toc = ru_books_reader.get_toc()
books_names = books_toc['header_0'].unique()

books_names
```

```python
book_ids = books_toc.axes[0].to_list()
```

```python
corpus_writer = CorpusWriter(Path("./files/new.zip"))
corpus_writer.add_fragment(ru_books_reader.get_frames(uids=book_ids).first())
corpus_writer.finalize()
```

```python
corpus_reader = CorpusReader(Path("./files/new.zip"))
corpus_reader.get_frames().first()
```

```python
for books_name in books_names:
    cleared_name = books_name.replace(" / ", " ")
    corpus_writer = CorpusWriter(Path(f"./files/ru_books/{cleared_name}.zip"))
    book_ids = books_toc[books_toc['header_0'] == books_name].axes[0].to_list()

    frames = ru_books_reader.get_frames(uids=book_ids).to_list()
    for frame in frames:
        corpus_writer.add_fragment(frame, frame['file_id'].values[0])
    corpus_writer.finalize()
```
