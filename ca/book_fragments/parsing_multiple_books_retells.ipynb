```python
from grammar_ru.corpus.corpus_writer import CorpusWriter

books_paths = Path('../files/en_books').glob('*')
incorrect_books = []

for book_path in books_paths:
    book_name = book_path.name[:-4]

    book_reader = CorpusReader(Path(book_path))
    book_retell = CorpusReader(Path(f"./files/en_corpus_retells/{book_name}.retell.zip"))

    book_retell_ids = set(book_retell.get_toc().index.array)

    writer = CorpusWriter(Path(f"../files/en_cleared_books/{book_name}.zip"))

    for frame in book_reader.get_frames():
        frame_id = frame.file_id.values[0]

        if frame_id in book_retell_ids:
            writer.add_fragment(frame)
    
    writer.finalize()
```

```python
from ca.utils_translate import translate_subcorpus

PARALLEL_CORPUS_PATH = Path('./files/parallel_corpus.zip')
if PARALLEL_CORPUS_PATH.is_file():
    os.unlink(PARALLEL_CORPUS_PATH)

def add_relation(df_1,df_2,name_1,name_2):
    rel_1 = pd.DataFrame({'file_1':df_1, 'file_2':df_2,'relation_name':f"{name_1}_{name_2}"})
    rel_2 = pd.DataFrame({'file_1':df_2, 'file_2':df_1,'relation_name':f"{name_2}_{name_1}"})
    rel = pd.concat([rel_1,rel_2])
    return rel
```

```python
books_paths = Path('../files/en_cleared_books').glob('*')
incorrect_books = []

for book_path in books_paths:
    book_name = book_path.name[:-4]

    book_reader = CorpusReader(Path(book_path))
    book_retell = CorpusReader(Path(f"./files/en_corpus_retells/{book_name}.retell.zip"))

    CorpusBuilder.update_parallel_data(
        PARALLEL_CORPUS_PATH,
        book_reader,
        f"en_book"
    )

    CorpusBuilder.update_parallel_data(
        PARALLEL_CORPUS_PATH,
        book_retell,
        f"en_retell",
        add_relation(book_reader.get_toc().index, book_retell.get_toc().index, f"en_book", f"en_retell")
    )
```
